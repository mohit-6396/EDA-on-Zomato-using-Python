# Visualizations

```python
zt["Country"].value_counts().index
```

```python
plt.figure(figsize = (15,6))
fig = plt.bar(zt["Country"].value_counts().index[:10], zt["Country"].value_counts().values[:10], color = "maroon")
plt.xlabel("Country", size = 15, color = "black")
plt.ylabel("No. of Restaurants", size = 15, color = "black")
plt.title("No. of Restaurants in a Country", size = 20, color = "black");
```

```python
# Distribution of restaurants having table bookings or not
zc1 = zt.copy()
zc1["Has Table booking"].replace({"No": 0, "Yes": 1}, inplace = True)
zc1["Has Table booking"].value_counts()
```

```python
fig = plt.pie(zc1["Has Table booking"].value_counts(), labels = zc1["Has Table booking"].replace({0 : "No", 1 : "Yes"}).unique(),
             autopct= "%0.2f%%")
plt.title("Distribution of Restaurants having Table Booking", color = "Black");
```

```python
# Distribution of restaurants having online delivery
colors = sns.color_palette("rocket")
fig = plt.pie(zt["Has Online delivery"].replace({"Yes" : 1, "No" : 0}).value_counts(), colors = ["sandybrown", "blueviolet"],
             labels = zt["Has Online delivery"].unique(), autopct = "%0.2f%%")
plt.title("Distribution of Restaurants having Online Delivery", color = "black");
```

```python
zt.groupby("Country")["Aggregate rating"].mean().values
```

```python
# Countries according to the average aggregate ratings of all their restaurants
plt.figure(figsize = (18,8))
country_name = zt.groupby("Country")["Aggregate rating"].mean().index
country_rating = zt.groupby("Country")["Aggregate rating"].mean().values
fig = sns.barplot(data = zt, x = country_name,
                  y = country_rating)
plt.xlabel("Country", color = "Black", size = 20)
plt.ylabel("Average Rating", color = "Black", size = 20)
plt.title("Countries according to the avg. ratings of Restaurants", color = "Black", size = 20)
for i in fig.containers:
    fig.bar_label(i,)
```
